```python
import pandas as pd

# Load datasets
movies = pd.read_csv("/content/movies_metadata.csv", low_memory=False)
credits = pd.read_csv("/content/credits.csv")
keywords = pd.read_csv("/content/keywords.csv")
links = pd.read_csv("/content/links.csv")

# Convert IDs to numeric
movies['id'] = pd.to_numeric(movies['id'], errors='coerce')
credits['id'] = pd.to_numeric(credits['id'], errors='coerce')
keywords['id'] = pd.to_numeric(keywords['id'], errors='coerce')
links['tmdbId'] = pd.to_numeric(links['tmdbId'], errors='coerce')
valid_movie_ids = set(links['tmdbId'].dropna())
movies = movies[movies['id'].isin(valid_movie_ids)]

# Merge datasets
master_df = movies.merge(credits, on='id', how='left')
master_df = master_df.merge(keywords, on='id', how='left')

#  Remove the extra invalid row
master_df = master_df.dropna(subset=['cast', 'crew', 'keywords'])
#  combined NLP feature
master_df['combined_text'] = (
    master_df['overview'].fillna('') + ' ' +
    master_df['genres'].fillna('') + ' ' +
    master_df['keywords'].fillna('') + ' ' +
    master_df['cast'].fillna('')
)

print(master_df.columns)
print(master_df.shape)
```
